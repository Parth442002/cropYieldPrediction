```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
data=pd.read_csv("../data/Final.csv")
data.head(10)
```

##### Dropping Year Column

```python
data.drop("Year",inplace=True,axis=1)
data.head(10)
```

Columns Which require OneHot-Incoding:-
+ Year
+ Item

```python
data=pd.get_dummies(data,columns=["Area","Item"])
```

Columns Which Need Scaling
+ Pest
+ RainFall
+ Temperature

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
cols = ['Pest','Rainfall',"Temp"]
data[cols]= sc.fit_transform(data[cols])
data.head()
```
