```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/temp.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
null_rows = df.isnull().any(axis=1)
print(df[null_rows])
```

```python
# removing the null values
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
# resetting the index
df = df.reset_index(drop=True)
```

```python
df.head(20)
```

```python
df.tail()
```

```python
sns.lineplot(x='year', y='avg_temp', data=df)
plt.xlabel('Year')
plt.ylabel('Average Temperature')
plt.show()
```

```python

plt.figure(figsize=(30, 10)) # Adjust the figure size as per your requirement
sns.boxplot(x='country', y='avg_temp', data=df)
plt.xlabel('Country')
plt.ylabel('Average Temperature')
plt.xticks(rotation=90)
plt.show()

```
