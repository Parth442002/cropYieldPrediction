```python
df.drop(['Unit'],axis=1,inplace=True)
df.head()
```

```python
sns.lineplot(x="Year", y="Value", hue="Area", data=df)
plt.xlabel('Year')
plt.ylabel('Pesticide Use')
plt.title('Pesticide Use by Country Over Time')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(50, 5))
sns.barplot(x="Area", y="Value", data=df, estimator=sum)
plt.xticks(rotation=90)
plt.xlabel('Country')
plt.ylabel('Total Pesticide Use')
plt.title('Total Pesticide Use by Country')
plt.show()

```
