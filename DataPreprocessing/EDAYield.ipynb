# Handling YieldCSV

```python
yieldCSV.head(10)
```

```python
### Compareing Domain Code and Domain Name
domain_code_len=len(pd.unique(yieldCSV['Domain Code']))
domain_len=len(pd.unique(yieldCSV["Domain"]))
print(domain_code_len==domain_len)
```

```python
yieldCSV["Domain Code"].unique()
```

```python
yieldCSV["Domain"].unique()
```

### We can infer that Domain Code and Domain is the same information and hence we can drop one of them.

```python
## Droppping Domain Code
yieldCSV.drop(["Domain Code"],axis=1,inplace=True)
yieldCSV.head(10)
```

```python
### Comapring Area Code and Area
area_code_len=len(pd.unique(yieldCSV['Area Code']))
area_len=len(pd.unique(yieldCSV["Area"]))
print(area_code_len==area_len)
```

```python
print(len(yieldCSV["Area Code"].unique()))
print(len(yieldCSV["Area"].unique()))
```

```python
### Creating A dict to store area_code to area mapping
area_=yieldCSV.iloc[:,2]
area_code=yieldCSV.iloc[:,3]
areaMapping = pd.merge(area_,area_code, right_index = True,
               left_index = True)
areaMapping.drop_duplicates(inplace=True)
areaMapping.reset_index()
areaMapping.head(10)
```

### Saving the Area and Area Code Mapping in a csv

```python
areaMapping.to_csv("../Helpers/mappings/areaMapping.csv",index=False)
```

```python
### Removing Area Code 
yieldCSV.drop(["Area Code"],axis=1,inplace=True)
```
